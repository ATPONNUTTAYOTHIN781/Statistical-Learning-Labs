## 1. For the Naive Bayes model, use grid search 5-fold cross-validation across different values of `alpha` to find the best model.

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import f1_score
from scipy.stats import uniform
```

```python
# สร้าง Pipeline
pipeline_grid = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('nb', MultinomialNB())
])

# กำหนดค่าที่จะทดสอบใน Grid Search
param_grid = {
    'nb__alpha': [0.001, 0.01, 0.1, 0.5, 1.0, 5.0, 10.0]
}

# สร้าง Grid Search ด้วย 5-fold cross-validation
grid_search = GridSearchCV(pipeline_grid, param_grid, cv=5, scoring='f1_macro')
grid_search.fit(Xtrain, ytrain)
```

```python
# สร้าง Pipeline เหมือนเดิม
pipeline_random = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('nb', MultinomialNB())
])

# กำหนดการสุ่ม alpha จาก uniform distribution
param_dist = {
    'nb__alpha': uniform(loc=0.001, scale=10)
}

# สร้าง Randomized Search
random_search = RandomizedSearchCV(
    pipeline_random,
    param_distributions=param_dist,
    n_iter=20,
    cv=5,
    scoring='f1_macro',
    random_state=42
)

random_search.fit(Xtrain, ytrain)

y_pred_random = random_search.predict(Xtest)
f1_random = f1_score(ytest, y_pred_random, average='macro')

print("\n📗 [ผลลัพธ์จาก Random Search]")
print("ค่าที่ดีที่สุดของ alpha:", random_search.best_params_['nb__alpha'])
print("ค่า f1_macro บน test set:", round(f1_random, 4))
```

```python
print("\nเปรียบเทียบผลลัพธ์:")
print(f"- Grid Search: alpha = {grid_search.best_params_['nb__alpha']}, f1_macro = {round(f1_grid, 4)}")
print(f"- Random Search: alpha = {random_search.best_params_['nb__alpha']}, f1_macro = {round(f1_random, 4)}")

if f1_random > f1_grid:
    print("Random Search ได้คะแนนดีกว่า")
elif f1_random < f1_grid:
    print("Grid Search ได้คะแนนดีกว่า")
else:
    print(" คะแนนเท่ากัน")
```
