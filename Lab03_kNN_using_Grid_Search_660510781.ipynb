```python
from sklearn.datasets import fetch_20newsgroups

train = fetch_20newsgroups(subset='train')
test = fetch_20newsgroups(subset='test')

Xtrain = train.data[:3000]
ytrain = train.target[:3000]
Xtest = test.data[:500]
ytest = test.target[:500]

print("X:", len(Xtest))
print("y:", len(ytest))
```

```python
print("X[0]:", Xtrain[0])
print("y[0]:", ytrain[0])
```

```python
train.target_names
```

### Apply Tfidf ([documentation](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html))

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Define transformation
tfidf = TfidfVectorizer(max_features=5000, stop_words='english')

# fit+transform training set
Xtrain_tfidf = tfidf.fit_transform(Xtrain)

# แสดงขนาดของ matrix
print("Training set TF-IDF shape:", Xtrain_tfidf.shape)
```

```python
#tfidf.vocabulary_
```

### Classify with k-Nearest Neighbor (kNN) ([documentation](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html))

```python
from sklearn.neighbors import KNeighborsClassifier

# Define model
nb = KNeighborsClassifier(n_neighbors=5)

# Fit the model
nb.fit(Xtrain_tfidf, ytrain)
```

Evaluate on the test set using [`classification_report`](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html)

We will focus on the [F1-score](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.f1_score.html)

```python
from sklearn.metrics import classification_report

# Transform the test set
Xtest_tfidf = tfidf.transform(Xtest)

# Make predictions on the test set
ypred = nb.predict(Xtest_tfidf)

# report classification scores
print(classification_report(ytest, ypred, target_names=train.target_names))
```

### Combine all methods into a [pipeline](https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html)

```python
from sklearn.pipeline import Pipeline

# Combine all methods into a pipeline
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=5000, stop_words='english')),
    ('knn', KNeighborsClassifier(n_neighbors=5))
])

# Fit the pipeline to the training set
pipeline.fit(Xtrain, ytrain)

# Make predictions on the test set
ypred_pipeline = pipeline.predict(Xtest)

# report classification scores
print("\n=== Pipeline Classification Report ===")
print(classification_report(ytest, ypred_pipeline, target_names=train.target_names))
```

Now we will use [grid search cross-validation](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html) to find model with the best hyperparameters

![5CV](https://scikit-learn.org/stable/_images/grid_search_cross_validation.png)

```python
from sklearn.model_selection import GridSearchCV

params = {
    'tfidf__max_features': [1000, 5000],
    'tfidf__ngram_range': [(1, 1), (1, 2)],
    'knn__n_neighbors': [3, 5, 7]
}

# Define GridSearchCV
gridcv = GridSearchCV(pipeline, params,
                      scoring='f1_macro', cv=3)

# Fit and cross-validate the model on 3-fold data
gridcv.fit(Xtrain, ytrain)
```

```python
gridcv.best_estimator_
```

```python
# Make predictions on the test set
ypred_best = gridcv.predict(Xtest)

# Report classification scores
print("\nGrid Search CV Results:")
print(classification_report(ytest, ypred_best))
```

#### Exercise 2:

1. Use grid search 5-fold cross-validation across different values of the following two kNN parameters: `n_neighbors` and `metric`  **on the training set** to find the best model.

2. For the best value of `n_neighbors` and `metric` you found above, compute the `f1_macro` score **on the test set**.
* Print the value of `n_neighbors` and `metric`.
* Print the model's `f1_macro` score.

```python
from sklearn.metrics import f1_score
# Create a new pipeline for Exercise 2
pipeline_ex2 = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('knn', KNeighborsClassifier())
])
```

```python
# Define parameters for grid search focusing on n_neighbors and metric
params_ex2 = {
    'knn__n_neighbors': [3, 5, 7, 9, 11],
    'knn__metric': ['euclidean', 'manhattan', 'cosine', 'minkowski']
}
```

```python
# Define GridSearchCV with 5-fold cross-validation
gridcv_ex2 = GridSearchCV(pipeline_ex2, params_ex2,
                          scoring='f1_macro', cv=5,
                          n_jobs=-1, verbose=1)
```

```python
# Fit and cross-validate the model on training set
print("Performing 5-fold cross-validation...")
gridcv_ex2.fit(Xtrain, ytrain)

# Get the best parameters
best_n_neighbors = gridcv_ex2.best_params_['knn__n_neighbors']
best_metric = gridcv_ex2.best_params_['knn__metric']

print(f"\nBest n_neighbors: {best_n_neighbors}")
print(f"Best metric: {best_metric}")
```

```python
# Make predictions on test set with best model
ypred_ex2 = gridcv_ex2.predict(Xtest)

# Calculate f1_macro score on test set
f1_macro_score = f1_score(ytest, ypred_ex2, average='macro')

print(f"\nf1_macro score on test set: {f1_macro_score:.4f}")

print(f"\nBest cross-validation score: {gridcv_ex2.best_score_:.4f}")

# Show full classification report
print("\nFull classification report on test set:")
print(classification_report(ytest, ypred_ex2))
```
